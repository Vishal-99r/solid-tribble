WRITING A FUNCTION TO GET THE NUMBER OF MOVIES WITH START YEAR AND END

```python
def get_num_of_movies(director_name,start_year,end_year):
  director_ids_of_interest=directors.loc[directors['director_name']==director_name,'id']
  return len(movies.loc[(movies['director_id'].isin(director_ids_of_interest))&
           (movies['year']>=start_year)&
           (movies['year']<=end_year)
           ])
```

```python
get_num_of_movies("James Cameron",2022,2022)
```
